```python
#DFT Fungsi Sinus Cosinus
from matplotlib import pyplot as plt
import numpy as np
import math
import cmath
import array as arr
N = int(input("Jumlah Diskritisasi Fungsi: "))
t1 = float(input("Batas Bawah Rentang: "))
t2 = float(input("Batas Atas Rentang: "))

t = arr.array('f',[0])
f = arr.array('f',[0])
w = arr.array('f',[0])
g_real = arr.array('f',[0])
g_imaginer = arr.array('f',[0])
g = arr.array('f',[0])

t = [0]*(N+1)
f = [0]*(N+1)
w = [0]*(N+1)
g_real = [0]*(N+1)
g_imaginer = [0]*(N+1)
g = [0]*(N+1)

dt = (t2-t1)/N

for i in range (0,N,1):
  t[i]=t1+i*dt
  f[i]=np.sin(t[i])

plt.scatter(t, f)
plt.title("Sinyal Input")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.show()

#Discrete Fourier Transform

for i in range (0,N,1):
 w[i]=-(N//2)+(i)
 A=0
 B=0
 for j in range (0,N,1):
  A = A + f[j]*(np.cos(w[i]*t[j])*dt)  
  B = B - f[j]*(np.sin(w[i]*t[j])*dt)
 g_real[i]=A
 g_imaginer[i]=B
 g[i]=A*A+B*B

#plot norm
plt.scatter(w,g)
plt.xlim (t1,t2)
plt.xticks(np.arange(t1, t2+1, 1))
plt.stem(w,g)
plt.title("Sinyal Output")
plt.xlabel("w")
plt.ylabel("g(w)")
plt.show() 


#plot bagian real dan imaginer secara terpisah
plt.scatter(w,g_real,color='blue', marker='*', label='Real')
plt.scatter(w,g_imaginer,color= 'red', marker='.', label='Imaginer')
plt.xlim(-t1, t2)
plt.xticks(np.arange(t1, t2+1, 1))
plt.legend(loc='best')
plt.title("Sinyal Output Bagian Real dan Imaginer")
plt.xlabel("w")
plt.ylabel("g(w)")
plt.show()

#Part imaginer ada wiggle karena fungsi yang difourier transform adalah fungsi diskrit (dirac) dan fungsi yang ditransformasi fungsi diskontinu.
```
